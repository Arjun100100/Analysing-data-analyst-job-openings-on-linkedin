# LinkedIn Data Analyst Jobs Analysis

We will be working with data of 2 countries & a continent- USA, Canada and Africa. The contains information of Data Analyst Vacancies uploaded on LinedIn. We will primarly clean, model and merge the data in python, then visualise it in the for of Tableau Dashboard.

```python
#Libraries set up

import pandas as pd
import numpy as np

#import the datasets

africa_data= pd.read_csv('/content/linkedin-jobs-africa.csv')
canada_data = pd.read_csv('/content/linkedin-jobs-canada.csv')
usa_data= pd.read_csv('/content/linkedin-jobs-usa.csv')

print(canada_data.head(5))
```

```python
print(usa_data.head(5))
```

```python
print(africa_data.head(5))
```

In order to merge all datasets into one, we will need to create a common column 'Country'. In the Africa we will need to do some modeling techniques to extract country name from location

```python
usa_data['Country'] = "USA"
canada_data['Country'] = "Canada"

print(usa_data.columns)
print(canada_data.columns)
```

```python
sp_data = africa_data['location'].str.split(',',expand=True)
print(sp_data)
```

```python
sp_data[2].fillna(sp_data[1],inplace=True)
print(sp_data)
```

```python
sp_data[2].fillna(sp_data[0],inplace=True)
sp_data.head(5)
```

```python
sp_data.drop([0,1],axis=1,inplace=True)
```

```python
sp_data.head(5)
```

Now we have extracted the accurate country names in Africa, lets join it with original Africa_data

```python
africa_data['Country'] = sp_data[2]
```

```python
africa_data.head(5)
```

Now we will merge all 3 datasets together

```python
data = pd.concat([africa_data,usa_data,canada_data],axis=0).reset_index()
data.head(2)
```

There certain columns in data which we donot need, lets remove them now.

```python
data.drop(['index','description','criteria','link'],axis=1,inplace=True)
data.columns
```

```python
data.info()
```

Now we have two problems to deal with before starting visualisation.
1. Null values in Salary
2. Data type on columns

First we will deal with null values, by filing them mean of salary. But before there is lot to clean in salary column

```python
from functools import reduce
import re
data["salary"].value_counts()
string="$50.00\r\n            -\r\n            $60.00"
def clean_sal(text):
    if type(text)!=float:
        list_nums=re.findall("[-+]?\d*\.*\d+",text)
        return float(reduce(lambda a,b: (a+b)/2,list(map(float,list_nums))))
#clean_sal(string)
data["Compensation"]=data["salary"].apply(clean_sal)
data.drop(columns=["salary"],inplace=True)
```

```python
data.groupby(['Country']).sum(['Compensation'])
```

As we can check from above, Africa continent data has no salary info. In real life, one should share this issue with higher authority and try to get some data. So here we will fill nan with randome value between min and max compenation.

```python
data['Compensation'].fillna(value=pd.Series(np.random.uniform(20, 490, len(data['Compensation']))),inplace=True)
data['Compensation']=round(data['Compensation'],3)
data.head(5)
```

```python
data.info()
```

Now we need to change the data type of posted_ date columns.

```python
data['posted_date']=pd.to_datetime(data['posted_date'])
data.info()
```

Finally, lets save the data frame as CSV for Tableau

```python
data.to_csv('combined_jobs_data.csv')
```
